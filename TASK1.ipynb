```python
import pandas as pd
```

**DATA LOADING**

```python
stud=pd.read_csv("student-mat.csv",sep=";")
print(stud)
print(stud.head())
print(stud.columns)
```

**DATA EXPLORATION**

```python
stud.isnull().sum()
```

```python
stud.dtypes
```

```python
stud.shape
```

**Data cleasning**

```python
stud.dropna(inplace=True)
stud.isnull().mean()
```

```python
d1=stud.drop_duplicates()
d1
```

**DATA ANALYSIS QUESTION**

```python
avg=stud['G3'].mean()
print("The average score in math (G3)",avg)
```

```python
ab=stud.loc[stud['G3']>15]
print("students scored above 15 in their final grade (G3)",ab)
```

```python
cr=stud['studytime'].corr(stud['G3'])
print("students scored above 15 in their final grade (G3)",cr)
```

```python
high_gen=stud.groupby(['sex'])['G3'].mean()
print("gender has a higher average final grade (G3)",high_gen)
```

**DATA VISUALIZATION**

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(bins=20,x=stud['G3'])
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(stud['G3'], stud['studytime'])
plt.xlabel('G3 (Final Grade)')
plt.ylabel('Study Time')
plt.title('Scatter Plot of Final Grade vs Study Time')
plt.show()
```

```python
avg_scores = stud.groupby('sex')['G3'].mean()

plt.bar(avg_scores.index, avg_scores.values, color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Average Scores by Gender')
plt.show()
```
